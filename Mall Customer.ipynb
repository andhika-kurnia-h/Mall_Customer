```python
# library
import pandas as pd
import numpy as np
```

```python
# Import dataset
dataset = pd.read_csv("Mall_Customers.csv")
dataset.head()
```

Explore dataset

```python
explore_data = dataset
```

```python
explore_data['Genre'].value_counts()
```

```python
explore_data['Spending Score (1-100)'].plot(kind='kde')
```

Berdasarkan grafik tersebut dapat kita lihat density dengan range score 0-25, 25-75, dan 75-100.
Hal yang perlu di explore :
1. Bagaimana relasi antara spending score dengan annual income (spending score dibuat 3 kategori)

```python
explore_data.loc[explore_data['Spending Score (1-100)'] <= 25, 'Define Spend'] = 'Low'
explore_data.loc[explore_data['Spending Score (1-100)'] > 25, 'Define Spend'] = 'Middle'
explore_data.loc[explore_data['Spending Score (1-100)'] > 75, 'Define Spend'] = 'High'
```

```python
explore_data.groupby('Define Spend')['Annual Income (k$)'].plot(kind='kde')
```
